### Step 3: Train/Test split

```python
#Sci-kit learn --> python library
```

```python
from sklearn.model_selection import train_test_split
#This line is used to import the function that splits the dataset into training and testing sets from scikit learn 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split( Independent, Dependent, test_size = 0.30, random_state =0) 
#This line splits entire datasets into four parts like X_train, X_test, Y_train, Y_test .test_size = 0.30 --->allocating 30% (0.30) of the data for testing,The remaining 70% (0.70) of the data is used to train the model.random_state = 0 --->  ensure that every time the code runs, the data is split in the same way.
```

### Step 4:Importing Algorithm and Model Learning 

```python
from sklearn.linear_model import LinearRegression
#This line imports the LinearRegression model from the Scikit-learn library.
```

```python
regressor = LinearRegression()
#This line create a Linear Regression model and store it in a variable called Regressor
```

```python
regressor
#This line shows the Linear Regression model object and confirms that the model is created.
```

```python
regressor.fit(X_train, Y_train)
#This line trains the Linear Regression model using the training input data (`X_train`) and training output data (`Y_train`).
```

### Step 5: To Find Weight and Bias

```python
Weight = regressor.coef_
#This line extracts the weight (coefficient) learned by the Linear Regression model and stores it in the variable Weight.
```

```python
Bias = regressor.intercept_
#This line extracts the bias (intercept) learned by the Linear Regression model and stores it in the variable bias.
```

```python
Weight
#This line displays the weight value of the Linear Regression model.
```

```python
Bias
#This line displays the bias value of the Linear Regression model.
```

```python
Y_pred = regressor.predict(X_test)
#This line uses the trained Linear Regression model to predict the output values for the testing input data.
```

```python
Y_pred
#This line displays the predicted output values generated by the model for the test dataset.
```

### Step 6: To Find The Accuracy Score 

```python
from sklearn.metrics import r2_score
#This line imports the r2_score function from Scikit-learn to measure the accuracy of the regression model.
```

```python
r_score = r2_score(Y_test, Y_pred)
#This line calculates the R² (R-squared) score by comparing the actual output values (Y_test) with the predicted values (Y_pred).
```

```python
r_score
#This line displays the accuracy score of the model.
```

### Step 7: To Save The Model

```python
import pickle 
#This line imports the pickle library, which is used to save and load machine learning models.
```

```python
filename = "Final_model_SLR.pkl"
#This line assigns the name of the file used to save the trained machine learning model.
```

```python
pickle.dump(regressor,open(filename,'wb'))
#This line saves the trained Linear Regression model (regressor) into a file using the pickle library in write-binary mode, so it can be reused later without retraining.
```

```python
loaded_model = pickle.load(open('Final_model_SLR.pkl','rb'))
#This line loads the saved machine learning model from the file Final_model_SLR.pkl using the pickle library so it can be used again for prediction.
```

```python
a = int(input("Enter Your Years Of experience:"))
#Takes years of experience from the user and converts it into an integer.
result=loaded_model.predict([[a]])
# Uses the trained (loaded) model to predict salary based on the given experience.
print("Predicted salary based on your Experience is : ",result)
#Prints the predicted salary to the user.
```
